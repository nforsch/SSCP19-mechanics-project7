```python
from pyDOE import *
from scipy.stats.distributions import norm
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

## Latin Hypercube Sampling

see: https://pythonhosted.org/pyDOE/randomized.html
        
> lhs(n, [samples, criterion, iterations])

### Generate samples and normalize

```python
n = 4
samples = 100
X = lhs(n, samples=samples)
means = [1, 2, 3, 4]
stdvs = [0.1, 0.5, 1, 0.25]
for i in range(4):
    X[:, i] = norm(loc=means[i], scale=stdvs[i]).ppf(X[:, i])
```

### Plot histogram

```python
fig,axs = plt.subplots(1,4,figsize=(15,4))

for count,ax in enumerate(axs):
    ax.hist(X[:,count],bins=int(round(samples*0.1)),density=True)
    x = np.linspace(means[count] - 3*stdvs[count], means[count] + 3*stdvs[count], 100)
    ax.plot(x, stats.norm.pdf(x, means[count], stdvs[count]),LineWidth=4)
    ax.set_xlabel("Variable "+str(count))
```
